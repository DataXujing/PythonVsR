## 前言

    在数据分析、数据挖掘中，相同模型有不同的版本，比如Xgboost有R版本，Python版本和Java版本（当然还有C版本），前期我们曾做过实验，相同的模型在R和Python中的表现是不同的，在前期的一些大数据分析挖掘比赛中，我们曾做过使用不同版本的相同xgboost，RF，SVM,GBDT进行数据的预测。这种模型的融合要优于单版本的融合（AUC大约高出0.3%）。在企业分析中，由于数据分析师所擅长的分析工具不同，在协同开发中会出现不同工具的交叉合作。综上，我们需要融合不同开发工具，实际上这也是数据分析师所必须面临的问题：一个合格的数据分析师，需要掌握多种分析工具。

![](order1.png)

    实际上不仅在数据科学中存在这些问题：
    
+ Windows or Linux or Mac
+ IPhone or Android
+ R or Python
    
    

    为什么呢？简单看看Python的数据分析能力：
    
+ NumPy/Scipy 矩阵处理和科学计算

+ pandas 结构化表格处理

+ statsmodels 统计分析统计模型

+ scikit-learn 机器学习

+ urllib2/requests 网页数据抓取

+ BeautifulSoup 网页结构数据解析

+ theano/keras 基于神经网络的深度学习

    两种工具的介绍：差异点
    
+ Python是一种通用编程语言，R偏向于统计专业

+ R有更丰富的统计分析函数，Python长于机器学习

+ R有更好的可视化包，Python正在进步

+ Python和R的核心算法非常简洁，R包的语法兼收并蓄，错综复杂

#### 先研究一下网页结构

```python
from IPython.display import IFrame

IFrame("http://t.dianping.com/list/qingdao-category_19?pageIndex=1",width="100%",height=500)
```

#### 加载需要的模块

```python
# -*- coding: utf-8 -*-
"""
Created on Tue Jul 10 21:41:06 2017

@author: xujing
"""

import requests

from bs4 import BeautifulSoup

import re

#url="http://t.dianping.com/list/qingdao-category_19"
#r=requests.get(url)
#soup_packtpage=BeautifulSoup(r.text)

```

#### 找到网页中需要的内容

```python
def namefunc(qdXiCan):
    names=[x.findChild("h3",attrs=None).string for x in qdXiCan]
    return names

def newprice(qdXiCan):
    newpri=[float(str(x.findChild('em').string)) for x in qdXiCan]
    return newpri

def oldprice(qdXiCan):
    oldpri=[x.findChild('del',attrs=None).string for x in qdXiCan]
    return oldpri

def getinfo(url):
    r=requests.get(url)
    soup_packtpage=BeautifulSoup(r.text)
    
    qdXiCan=soup_packtpage.findAll("li",attrs={'class':'tg-floor-item'})
    names=namefunc(qdXiCan)
    newpri=newprice(qdXiCan)
    oldpri=oldprice(qdXiCan)
    
    TuanGou={'names':names,'newpri':newpri,'oldpri':oldpri}
    return TuanGou
```

#### 批量处理网页

```python
urls=[]
index=range(0,51)

for x in index:
    urls.append('http://t.dianping.com/list/qingdao-category_19?pageIndex='+str(x))
    urls

TuanGou={'names':[],'newpri':[],'oldpri':[]}

for url in urls:
    new=getinfo(url)
    TuanGou['names'].extend(new['names'])
    TuanGou['newpri'].extend(new['newpri'])
    TuanGou['oldpri'].extend(new['oldpri'])

#这里要注意extend和append的区别，两个函数均仅可接收一个参数
#extend 接受一个参数，这个参数总是一个 list，并且把这个 list 中的每个元素添加到原 list 中
#append 接受一个参数，这个参数可以是任何数据类型，并且简单地追加到 list 的尾部
```

#### 抓取结果展示

```python
import pandas as pd
df =pd.DataFrame(TuanGou)

df.head()
```

#### 把数据扔给R进行画图

```python
%%R -i df -w 500 -h 300

library(ggplot2)
library(dplyr)

p1<-df%>%mutate(index=1:nrow(df))%>%ggplot(.,aes(x=index,y=oldpri))+geom_point(size=4,alpha=0.5,color="green")+geom_smooth(aes(x=index,y=oldpri))

p2<-p1+geom_point(aes(x=index,y=newpri),size=3,alpha=0.4,color="red")+geom_smooth(aes(x=index,y=newpri))+
labs(title="DaZhongDianPing(QD)",x="XiCan",y="Price(RMB)")+theme(plot.title=element_text(hjust=0.5))

p2

#TMD中国汉子竟然不显示，修改了配置文件也不行
```

**用R做线性回归**

```python
rscript='''
df=read.csv("iris.csv")
m=lm(Sepal_Length~Sepal_Width,data=df)
'''

pymodel=robjects.r(rscript)

print pymodel.names
```

```python
import numpy as np
print np.mean(pymodel.rx2("residuals"))
```

```python
import rpy2.robjects as robjects
rscript='''
df=read.csv("iris.csv")
m=lm(Sepal_Length~Sepal_Width,data=df)
'''
res=robjects.r("summary(m)")

print res
```

```python
print res.names
```

```python
print res.rx2("coefficients")
```

```python
print np.mat(res.rx2("coefficients"))

np.mat(res.rx2('coefficients'))[1,3]
```
